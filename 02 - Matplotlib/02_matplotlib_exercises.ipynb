### Step 1
- Import pandas as pd
- Then import matplotlib.pyplot as plt
- Then set the %matplotlib inline
- Set the figure.figsize parameter to (5,2) so that it fits nicely on the screen.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt


fig = plt.figure(figsize=(5, 2))
```

### Step 2
- Read the data from CSV file stock-data.csv
- Use the index_col=0
- Also, it is convenient to use parse_dates=True
- Take head of data

```python
d0 = pd.read_csv("stock_data.csv",index_col=0, parse_dates=True)
```

```python
d0.head()
```

### Step 3
- make a plot of Close price.

```python
d0["Close"].plot()
```

### Step 5
- Make a figure and axes by calling matplotlib.pyplot from plt
- Now make a chart with Closing price

```python
fig, ax = plt.subplots()
d0["Close"].plot(ax=ax)
```

### Step 6
- Make new figure of 4 subplots in 2 rows and 2 columns
- Make charts of Open, High, Low, and Close in each axes

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

d0['Open'].plot(ax=axes[0,0])

d0['High'].plot(ax=axes[0,1])

d0['Low'].plot(ax=axes[1,0])

d0['Close'].plot(ax=axes[1,1])

plt.tight_layout()
```
