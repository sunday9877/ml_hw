### Step 1
- import the Pandas library
- read the **gdp-00s.csv** file into a DataFrame **gdp00**
- ensure to set the **index_col** to 0
- Call **gdp00.head()**

```python
import pandas as pd
```

```python
gdp00 = pd.read_csv("gdp-00s.csv",index_col=0)
```

```python
gdp00.head()
```

### Step 2
- read the **gdp-10s.csv** file into a DataFrame **gdp10**
- ensure to set the **index_col** to 0
- Call **gdp10.head()**

```python
gdp10 = pd.read_csv("gdp-10s.csv",index_col=0)
```

```python
gdp10.head()
```

### Step 3
- concatenate the DataFrames **gdp00** and **gdp10** into **gdp** 
- HINT: concatenate() exists in the pandas library.
- HINT: **axis=1** parameter 
- Call **gdp.head()**

```python
gdp0 = pd.concat([gdp00, gdp10], axis=1)
gdp = gdp0.rename_axis(None, axis=0)
```

```python
gdp.head()
```

### Step 4
- Drop all rows with **NaN**
- use argument **inplace=True**
- Call **gdp.head()**
- HINT: dropna() method on the DataFrame object

```python
gdp.dropna(inplace=True)
```

```python
gdp.head()
```

### Step 5
- Calculate the GDP growth from 2000-2019 as a percentage in a column named **19y-pct**
- Call **gdp.head()**

```python
gdp["19y-pct"] = (gdp["2019"] - gdp["2000"])/gdp["2000"]*100
```

```python
gdp.head()
```

### Step 6
- Calculate 10y-pct from 2009-2019
- Head

```python
gdp["10y-pct"] =  (gdp["2019"] - gdp["2009"])/gdp["2009"]*100
```

```python
gdp.head()
```

### Step 7
- List the 10 countries with highest GDP growth rate in the 19 years 2000-2019.
- HINT:
    - sort_value
    - ascending=False
    - head(10)

```python
gdp_highest19 = gdp.sort_values(by=["19y-pct"],ascending=False)
```

```python
gdp_highest19.head(10)
```

### Step 8
- List the 10 countries with highest GDP growth rate between 2000-2019.
- HINT:
    - sort_value
    - ascending=False
    - head(10)

```python
gdp_highest10 = gdp.sort_values(by=["10y-pct"],ascending=False)
```

```python
gdp_highest10.head(10)
```

### Step 9
- Export data to Excel
- Name the file **gdp.xlsx**

```python
gdp.to_excel("gdp.xlsx")
```
